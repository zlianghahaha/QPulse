```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# X: parametric pulse id
# Y: Entangling Capability
```

```python
def get_list(a, b, seq):
    if seq == 'compare':
        return np.array([[a[i], b[i]] for i in range(6) ]).reshape(12)
    elif seq == 'seq': 
        return np.array(a + b)
```

```python
x = range(1, 13)
```

```python
qubit_2_temp = [0.1097, 0.0967, 0.1074, 0.0911, 0.0890, 0.0862]
qubit_2_rand = [0.3002, 0.1648, 0.1143, 0.0947, 0.0946, 0.0830]
```

```python
qubit_3_temp = [0.0920, 0.07866, np.nan, np.nan, 0.05634, 0.05223]
qubit_3_rand = [0.1059, 0.1000, np.nan, np.nan, 0.0911, 0.1422]
y_3 = get_list(qubit_3_temp, qubit_3_rand, 'seq')
y3_not_nan_inds = list(filter(lambda x: ~np.isnan(y_3[x]), range(12)))
y3_not_nan_inds
```

```python
qubit_4_temp = [0.1680, 0.0744, 0.1171, 0.1174,  0.0707, 0.0565]
qubit_4_rand = [1.0164, 1.5209, 0.3414, 0.1748, 0.4051, 0.2395]
```

```python
fig, ax = plt.subplots(figsize=(6,3))
markersize = 10
markerwidth = 1.5
ax.plot(x, get_list(qubit_2_temp, qubit_2_rand, 'seq'), 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='2q')
ax.plot([x[i] for i in y3_not_nan_inds], [y_3[i] for i in y3_not_nan_inds], 'X', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='3q')
ax.plot(x, get_list(qubit_4_temp, qubit_4_rand, 'seq'), 'o', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='4q')
ax.legend()
ax.set_xlabel("Pulse ID")
ax.set_ylabel("Expr")
ax.grid(linestyle=':')
plt.xticks(x, x)
```

```python
fig, ax = plt.subplots(figsize=(12,2), dpi = 400)
markersize = 10
markerwidth = 1.5
ax.plot(x, get_list(qubit_2_temp, qubit_2_rand, 'seq'), 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='2q')
ax.plot([x[i] for i in y3_not_nan_inds], [y_3[i] for i in y3_not_nan_inds], 'X', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='3q')
ax.plot(x, get_list(qubit_4_temp, qubit_4_rand, 'seq'), 'o', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='4q')
ax.legend()
ax.set_title('Expressivity for Pulse Sets',fontsize = 14)
ax.set_xlabel("Pulse ID", fontsize = 14)
ax.set_ylabel("Expr",fontsize = 14)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(linestyle=':')
plt.xticks(x, x)

# 创建一个新的子图，并设置其范围
ax_zoom = fig.add_axes([0.16, 0.48, 0.35, 0.3])
ax_zoom.plot(x, get_list(qubit_2_temp, qubit_2_rand, 'seq'), 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='2q')
ax_zoom.plot([x[i] for i in y3_not_nan_inds], [y_3[i] for i in y3_not_nan_inds], 'X', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='3q')
ax_zoom.plot(x, get_list(qubit_4_temp, qubit_4_rand, 'seq'), 'o', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='4q')
ax_zoom.set_xlim(0, 6.5)  # 设置子图的 x 轴范围
ax_zoom.set_ylim(0.04, 0.20)  # 设置子图的 y 轴范围
ax_zoom.grid(linestyle=':')
ax_zoom.tick_params(axis='both', which='major', labelsize=12)  
rect = plt.Rectangle((0.5, 0.01), 6, 0.23, edgecolor='r', facecolor='none', linewidth=2)
ax.add_patch(rect)
xy = (3.5, 0.25)
xytext = (0.3, 0.5)
coordsA = 'data'
coordsB = 'axes fraction'
arrowprops = dict(facecolor='black', arrowstyle='->')
ax.annotate('', xy=xy, xytext=xytext, xycoords=coordsA, textcoords=coordsB, arrowprops=arrowprops)
#ax_zoom.annotate('', xy=xy, xytext=(0.2, 0.5), xycoords=coordsA, textcoords=coordsB, arrowprops=arrowprops)

plt.show()
```

```python
sub_x = range(1, 4)
expr_list = [
    [0.1680, 0.0491, 0.0211], #HE_4q_general
    [0.0744, 0.0205, 0.0204], #HE_4q_fixed
    [0.1171, 0.0523, 0.0227], #decay_4q_general
    [0.1174, 0.0296, 0.0137], #decay_4q_fixed
    [0.0707, 0.0544, 0.0296], #dressed_4q_general
    [0.0565, 0.0371, 0.0185] #dressed_4q_fixed
]
```

```python
for i in range(0,6):
  fig, ax = plt.subplots(figsize=(4,2), dpi = 400)
  markersize = 10
  markerwidth = 1.5
  for idx in range(len(sub_x)):
      ax.plot([sub_x[idx]], [expr_list[i][idx]], 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none',)
  ax.set_title('Pulse '+str(i+1), fontsize = 14)
  ax.set_xlabel("Number of layers", fontsize = 14)
  ax.set_ylabel("Expr", fontsize = 14)
  ax.tick_params(axis='both', which='major', labelsize=14)
  
  ax.grid(linestyle=':')
  plt.ylim(round(max(min(expr_list[i])-0.1, 0),2), round(max(expr_list[i])+0.05,2))
  ymin, ymax = plt.ylim()
  yticks = np.linspace(ymin, ymax, 3)
  plt.yticks(yticks, ["{:.2f}".format(y) for y in yticks])
    
```
