```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# X: parametric pulse id
# Y: Entangling Capability
```

```python
def get_list(a, b, seq):
    if seq == 'compare':
        return np.array([[a[i], b[i]] for i in range(6) ]).reshape(12)
    elif seq == 'seq': 
        return np.array(a + b)
```

```python
x = range(1, 13)
```

```python
qubit_2_temp = [0.107668, 0.196968, 0.120268, 0.190692, 0.114562, 0.183688]
qubit_2_rand = [0.189283, 0.32545, 0.152320, 0.228952, 0.151397, 0.240566]
```

```python
qubit_3_temp = [0.207315, 0.331858, np.nan, np.nan, 0.229003, 0.338206]
qubit_3_rand = [0.143932, 0.129466, np.nan, np.nan, 0.122544, 0.143901]
y_3 = get_list(qubit_3_temp, qubit_3_rand, 'seq')
y3_not_nan_inds = list(filter(lambda x: ~np.isnan(y_3[x]), range(12)))
y3_not_nan_inds
```

```python
qubit_4_temp = [0.196618, 0.340176, 0.226385, 0.341022, 0.216195, 0.333295]
qubit_4_rand = [0.04188, 0.08035, 0.091712, 0.162238, 0.111713, 0.189597]
```

```python
qubit_5_temp = [0.236962, 0.389713, np.nan, np.nan,0.264408, 0.382663]
qubit_5_rand = [0.160327, 0.245955, np.nan, np.nan,0.157715, 0.198773]
y_5 = get_list(qubit_5_temp, qubit_5_rand, 'seq')
y5_not_nan_inds = list(filter(lambda x: ~np.isnan(y_5[x]), range(12)))
y5_not_nan_inds
```

```python
fig, ax = plt.subplots(figsize=(6,3))
markersize = 10
markerwidth = 1.5
ax.plot(x, get_list(qubit_2_temp, qubit_2_rand, 'seq'), 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='2q')
ax.plot([x[i] for i in y3_not_nan_inds], [y_3[i] for i in y3_not_nan_inds], 'X', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='3q')
ax.plot(x, get_list(qubit_4_temp, qubit_4_rand, 'seq'), 'o', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='4q')
ax.plot([x[i] for i in y5_not_nan_inds], [y_5[i] for i in y5_not_nan_inds], 's', markersize=markersize ,markeredgewidth=markerwidth, markerfacecolor= 'none', label='5q')
ax.legend()
ax.set_xlabel("Pulse ID")
ax.set_ylabel("Ent")
ax.grid(linestyle=':')
plt.xticks(x, x)
```

```python
fig, ax = plt.subplots(figsize=(12,2), dpi = 400)
markersize = 10
markerwidth = 1.5
ax.plot(x, get_list(qubit_2_temp, qubit_2_rand, 'seq'), 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='2q')
ax.plot([x[i] for i in y3_not_nan_inds], [y_3[i] for i in y3_not_nan_inds], 'X', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='3q')
ax.plot(x, get_list(qubit_4_temp, qubit_4_rand, 'seq'), 'o', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none', label='4q')
ax.plot([x[i] for i in y5_not_nan_inds], [y_5[i] for i in y5_not_nan_inds], 's', markersize=markersize ,markeredgewidth=markerwidth, markerfacecolor= 'none', label='5q')
ax.legend()
ax.set_title('Entanglement Capability for Pulse Sets',fontsize = 14)
ax.set_xlabel("Pulse ID", fontsize = 14)
ax.set_ylabel("Ent",fontsize = 14)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(linestyle=':')
plt.xticks(x, x)
```

```python
ent_list = [
    [0.196618, 0.327831, 0.369223], #HE_4q_general
    [0.340176, 0.447405, 0.492115], #HE_4q_fixed
    [0.226385, 0.3476544, 0.425427], #decay_4q_general
    [0.341022, 0.444879, 0.527337], #decay_4q_fixed
    [0.216195, 0.326225, 0.383379], #dressed_4q_general
    [0.333295, 0.422594, 0.489424] #dressed_4q_fixed
]
```

```python
sub_x = range(1, 4)
```

```python
for i in range(0,6):
  fig, ax = plt.subplots(figsize=(4,2), dpi = 400)
  markersize = 10
  markerwidth = 1.5
  for idx in range(len(sub_x)):
      ax.plot([sub_x[idx]], [ent_list[i][idx]], 'd', markersize=markersize, markeredgewidth=markerwidth, markerfacecolor= 'none',)
  ax.set_title('Pulse '+str(i+1), fontsize = 14)
  ax.set_xlabel("Number of layers", fontsize = 14)
  ax.set_ylabel("Ent", fontsize = 14)
  ax.tick_params(axis='both', which='major', labelsize=14)
  
  ax.grid(linestyle=':')
  plt.ylim(round(max(min(ent_list[i])-0.1, 0),2), round(max(ent_list[i])+0.05,2))
  ymin, ymax = plt.ylim()
  yticks = np.linspace(ymin, ymax, 3)
  plt.yticks(yticks, ["{:.2f}".format(y) for y in yticks])
```
