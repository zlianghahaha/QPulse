```python
import matplotlib.pyplot as plt
import numpy as np

# Read data from the text file
with open('chemistry.txt', 'r') as f:
    data = f.readlines()

# Extract the columns as x, y1, and y2
x = []
y1 = []
y2 = []
for line in data:
    columns = line.split()
    y1.append(float(columns[-2]))
    y2.append(float(columns[-1]))

x1 = list(range(1, 1+len(y1)))
x = np.append(np.arange(1, 13), ['Two\nLocal', 'Real\nAmp'])
# Set up the grouped bar chart
x_indices = np.arange(len(x1))
bar_width = 0.25

fig, ax = plt.subplots(figsize=(13,2), dpi = 400)
bar1 = ax.bar(x_indices - bar_width / 2, y1, bar_width, label='H2', color = '#4460EF')
bar2 = ax.bar(x_indices + bar_width / 2, y2, bar_width, label='LiH', color = '#4BC9F1')



# Customize the chart
ax.set_xlabel('Pulse ID', fontsize=10)
ax.set_ylabel('Accuracy', fontsize=10)
ax.set_xticks(x_indices)
ax.set_xticklabels(x, fontsize=10)
ax.set_yticks(np.arange(0, 1.2, 0.25))
ax.set_ylim([0, 1])
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(True)
ax.legend(loc='upper center', fontsize=8)


# Show the chart
plt.show()
```
